**Function 1: pluralize_gift()**

**The gifts are listed in singular: for example, on day five the narrator receives “five golden rings”, but the entry in the data set for the gift on day five simply says “ring”.**

**Hint 1: The gifts on days six and nine have unusual pluralization. You may assume that in other data sets, there will be no additional special cases besides these types.**

```python
def pluralize_gift(gift):
  """
  Returns plural of a noun

  Parameters
  ----------
  gift: str
    A noun

  Return
  ------
  str
    Plural version
  """

  if gift.find("oo") != -1:
    gift = gift.replace("oo", "ee")
  elif gift.endswith("y") and gift[-2].lower() not in "aeiou":
    gift = gift[:-1] + "ies"
  else:
    gift = gift + "s"

  return gift
```

**Function 2: make_phrase()**

Write a function called make_phrase() that takes as input the necessary information, and returns a phrase. For example,

make_phrase(num_word = "ten",

            item = "lords",

            verb = "a-leaping",

            adjective = "",

            location = "")

should return

"ten lords a-leaping"

```python
import pandas as pd
def make_phrase(num, num_word, item, verb, adjective, location):
  """
  num: int
    The day number (e.g., 1, 5, 12).
  num_word: str
    The day number as a word (e.g., "one", "five", "twelve").
  item: str
    The gift item (e.g., "lord", "ring").
  verb: str or NaN
    The action of the gift (e.g., "a-leaping").
  adjective: str or NaN
    The adjective for the gift (e.g., "golden").
  location: str or NaN
    The location of the gift (e.g., "in a pear tree").

  Return
  """

  ## Step 1: Replace NAs with blank strings
  verb = "" if pd.isna(verb) else verb
  adjective = "" if pd.isna(adjective) else adjective
  location = "" if pd.isna(location) else location

  ## Step 2: If the day number is larger than 1, the gift items need pluralized!
  ### Hint: call the function you created above!
  if num > 1:
    item = pluralize_gift(item)

  # Initializing variable
  display_num_word = ""

  ## Step 3: Figure out if a gift item starts with a vowel
  if num == 1:
    # Check if the gift item starts with a vowel
    if item.lower().startswith(('a', 'e', 'i', 'o', 'u')):

  ## Step 4: For the first day, if the gift item starts with a vowel, replace the day with "an" and if the gift item does not start with a vowel, replace the day with "a" (e.g. a partridge in a pear tree). If it is not the first day, use just the number word (e.g. ten lords a leap)
      display_num_word = "an"
    else:
      display_num_word = "a"
  else:
    display_num_word = num_word

  ## Step 5: Put all of the pieces together into one string and return!
# I got some help from chat on this part because I wasn't quite sure how to merge, but this made a lot of sense after cross referncing online
  phrase_parts = [display_num_word, adjective, item, verb, location]

  # Filter out any empty strings from the list before joining
  non_empty_parts = [part for part in phrase_parts if part]

  return " ".join(non_empty_parts)
```

**Function 3: sing_day()**

Write a function called sing_day() that takes as input:

A dataset (input as a dataframe)

A number indicating which day to sing about (input as an integer)

The name of a column in the dataset that contains the phrases for each day (input as an tidy name)

For example,

sing_day(xmas, 2, Full.Phrase)
should return


On the second day of Christmas, my true love sent to me:

two turtle doves and

a partridge in a pear tree.

```python
def sing_day(dataset, num, phrase_col):
  """
  Sings a verse of the 12 Days of Christmas song.

  Parameters
  ----------
  dataset: pandas.DataFrame
    The dataset containing the song lyrics.
  num: int
    The day number to sing about.
  phrase_col: str
    The name of the column containing the full gift phrases.

  Return
  ------
  str
    The complete verse for the specified day.
  """

  # Step 1: Setup the intro line
  num_word = dataset.loc[dataset['Day'] == num, 'Day.in.Words'].iloc[0]
  intro = f"On the {num_word} day of Christmas, my true love sent to me:"

  # Step 2: Sing the gift phrases
  gifts = ""
  # Iterate through the gifts in reverse order for the current day
  for i in range(num, 0, -1):
    phrase_to_add = dataset.loc[i - 1, phrase_col]

    if i == 1:
        # Last gift line: add a period
        gifts += phrase_to_add + "."
    elif i == 2 and num > 1:
        # Second to last gift line (if not the first day): add comma, "and", and newline
        gifts += phrase_to_add + ", and\n"
    else:
        # Other gift lines: add a comma and a newline
        gifts += phrase_to_add + ",\n"


  # Step 3: Put it all together and return
  full_verse = intro + "\n" + gifts.strip() # Use strip() to remove trailing newline

  return full_verse
```

```python
print(sing_day(xmas, 2, "Full.Phrase"))
```

```python
# --- Challenge: Sing the Entire Song with Spacing ---
print("The full 12 days of chistmas song \n")
for day in range(1, 13):
  print(sing_day(xmas, day, "Full.Phrase"))
  # Add a blank space between the different verses of the song.
  print("\n")
```
