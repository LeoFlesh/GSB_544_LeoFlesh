```python
# I think one way I could clean it for easier data readability would be to get rid of spaces and replace them with _
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

#Another way, per the tips suggestion, is renaming the codes to what they mean -- had to search them up for this
df = df.rename(columns={
    "4046": "small_hass",
    "4225": "large_hass",
    "4770": "xlarge_hass"
})

# The main thing I need to clean is the who geography issue -- so I could find each unique value, then sort them into US, Region, State, or Metro based on where they fit into the data
import pandas as pd

# get all unique values
geo_vals = df["geography"].unique()

# Define known categories and what they could be named as
us_labels = {"total u.s.", "totalus", "total_us", "total u.s", "totalu.s."} # Added "totalu.s." and "total_u.s." for consistent matching after removing spaces
major_regions = {
    "west", "northeast", "southeast", "southcentral",
    "midsouth", "greatlakes", "plains", "westtex/newmexico"
}

# list of U.S. states for matching (lowercase) -- this I searched up because I didn't want to mannually put in each state one by one
from us import states
state_names = set([s.name.lower() for s in states.STATES])  # uses 'us' package

# Define classification logic
def classify_geo(name):
    n = str(name).strip().lower().replace(" ", "") # Remove spaces for consistent matching
    if n in us_labels:
        return "US"
    elif n in major_regions:
        return "Region"
    elif n in state_names:
        return "State"
    else:
        return "Metro"

df["geo_level"] = df["geography"].apply(classify_geo)

# The rest of this I used chaptgpt support to see how to check to make sure I did it correctly, because I went through the rest of the lab and was off on somethings
#before
print("Value counts for geo_level:")
print(df["geo_level"].value_counts())

print("\nUnique geography names classified as US:")
print(df[df["geo_level"] == "US"]["geography"].unique())

# Print unique geography names that, after processing, match major_regions
print("\nUnique geography names that match major_regions after processing:")
processed_geo_names = df["geography"].apply(lambda x: str(x).strip().lower().replace(" ", ""))
matching_region_names = df["geography"][processed_geo_names.isin(major_regions)].unique()
print(matching_region_names)
```

**8. The following plot shows, for all four California regions, the proportion of the average Hass avocado sales that are small, large, or extra large; conventional vs. organic. Recreate the plot; you do not have to replicate the exact finishing touches - e.g., color, theme - but your plot should resemble the content of this plot.**

```python
california_sales = df_california.groupby(["geography", "type"])[["small_hass", "large_hass", "xlarge_hass"]].sum().reset_index()
california_sales['total_hass_volume'] = california_sales['small_hass'] + california_sales['large_hass'] + california_sales['xlarge_hass']
display(california_sales)
```

```python
california_sales['small_hass_proportion'] = california_sales['small_hass'] / california_sales['total_hass_volume']
california_sales['large_hass_proportion'] = california_sales['large_hass'] / california_sales['total_hass_volume']
california_sales['xlarge_hass_proportion'] = california_sales['xlarge_hass'] / california_sales['total_hass_volume']
display(california_sales)
```

```python
california_proportions_melted = california_sales.melt(
    id_vars=['geography', 'type'],
    value_vars=['small_hass_proportion', 'large_hass_proportion', 'xlarge_hass_proportion'],
    var_name='hass_size',
    value_name='proportion'
)
display(california_proportions_melted)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
from matplotlib.patches import Patch # Import Patch for creating legend handles

# Create a copy of the melted proportions data
tbl = california_proportions_melted.copy()

# Define the desired order and labels for geographies, types, and sizes
geo_order  = ["Los Angeles", "Sacramento", "San Diego", "San Francisco"]
type_order = ["conventional", "organic"]
size_order = ["small_hass_proportion", "large_hass_proportion", "xlarge_hass_proportion"]
size_label = {"small_hass_proportion":"Small",
              "large_hass_proportion":"Large",
              "xlarge_hass_proportion":"Xlarge"}

# Filter the data to include only the specified California regions and types
tbl = tbl[tbl["geography"].isin(geo_order)].copy() #
tbl["geography"] = pd.Categorical(tbl["geography"], categories=geo_order, ordered=True)
tbl["type"] = tbl["type"].str.lower()
tbl = tbl[tbl["type"].isin(type_order)].copy() # Use .copy()
tbl["type"] = pd.Categorical(tbl["type"], categories=type_order, ordered=True)
tbl["hass_size"] = pd.Categorical(tbl["hass_size"], categories=size_order, ordered=True)

# DColors for each -- searched up on chat what codes would be close to the colors I wanted
color_for = {"Small":"#D97706", "Large":"#16A085", "Xlarge":"#7E6BBF"}

# Create a figure and a set of subplots (1 row, 2 columns), sharey=True ensures both subplots share the same y-axis
fig, axes = plt.subplots(1, 2, figsize=(12.5, 5), sharey=True)

# Loop through each avocado type (conventional and organic) to create a subplot
for ax, t in zip(axes, type_order):
    # Filter the data for the current type
    sub = tbl[tbl["type"] == t].copy() # Use .copy()

    # Pivot the data to a wide format suitable for stacking, index='geography' makes geography the rows, columns='hass_size' makes hass_size the columns, values='proportion' fills the table with proportion values
    # .reindex(geo_order) ensures the geographies are in the desired order
    p = sub.pivot(index="geography", columns="hass_size", values="proportion").reindex(geo_order)

    # Ensure all size columns exist in the pivoted data (fill with 0.0 if not present)
    for col in size_order:
        if col not in p.columns:
            p[col] = 0.0
    # Reindex columns to ensure they are in the desired size order for stacking
    p = p[size_order]

    # Create an array for the x-axis positions of the bars
    x = np.arange(len(p))
    # Initialize the 'bottom' of the stacked bars to zero
    bottom = np.zeros(len(p))

    # Loop through each size to create stacked bars
    for col in size_order:
        # Get the label and values for the current size
        lab = size_label[col]
        vals = p[col].to_numpy()
        # Add the bar for the current size on top of the previous sizes
        ax.bar(x, vals, bottom=bottom, label=lab, color=color_for[lab])
        # Update the 'bottom' for the next size
        bottom += vals

    # Add a title for the subplot (conventional or organic) with styling
    ax.set_title(t, bbox=dict(facecolor="#D0D0D0", edgecolor="none", pad=6), fontsize=12, weight="bold")

    # Set the x-axis ticks and labels
    ax.set_xticks(x)
    ax.set_xticklabels(p.index, rotation=25) # Rotate labels
    # Set the y-axis limits
    ax.set_ylim(0, 1.05)
    # Set the y-axis ticks
    ax.set_yticks([0, .25, .5, .75, 1.0])
    # Format the y-axis labels as percentages
    ax.yaxis.set_major_formatter(PercentFormatter(1.0))
    # Add a horizontal grid for better readability
    ax.grid(axis="y", linestyle="--", linewidth=0.6, alpha=0.6)
    # Hide the top and right spines (borders) of the plot
    for spine in ["top","right"]:
        ax.spines[spine].set_visible(False)
    # Set the y-axis label only for the left subplot
    if ax is axes[0]:
        ax.set_ylabel("Proportion")
    # Set the x-axis label for both subplots
    ax.set_xlabel("Region of California")

# Create custom legend handles
legend_handles = [
    Patch(facecolor=color_for["Xlarge"], label="Xlarge"),
    Patch(facecolor=color_for["Large"],  label="Large"),
    Patch(facecolor=color_for["Small"],  label="Small"),
]


fig.legend(
    legend_handles, ["Xlarge","Large","Small"], title="size",
    loc="center left", bbox_to_anchor=(0.905, 0.5), frameon=True,
    fancybox=True, edgecolor="#DDDDDD"
)

# Add a main title for the entire figure
fig.suptitle("Proportion of Average Hass Avocado Sales by Size", y=1.03, fontsize=14, weight="bold")
# Adjust the layout to prevent titles/labels from overlapping, leaving space for the legend
plt.tight_layout(rect=[0, 0, 0.88, 1])
# Display the plot
plt.show()
```
